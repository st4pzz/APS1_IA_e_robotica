# Cálculo do algoritmo Busca em Largura #

```python
import pandas as pd
import numpy as np
from main import largura
data_largura = largura()
df = pd.DataFrame(data_largura)
df
```

# Conversão do formato do tempo de processamento para micro segundos #

```python
lista_tempos = []
for tempos in data_largura["Tempo de Processamento"]:
    if tempos != "...":
        lista_tempos.append(tempos.microseconds)
    else:
        lista_tempos.append(tempos)
```

```python
df_final = {
    "Algoritmo": data_largura["Algoritmo"],
    "Objetivo": data_largura["Objetivo"],
    "Tempo de Processamento": lista_tempos
}

df_final = pd.DataFrame(df_final)
df_final.to_csv("largura.csv")
```

# Gráfico de tempo de processamento em função do número objetivo do Busca em Largura #

```python
#plotar gráfico de linha com o tempo de processamento em função do objetivo
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.lineplot(x="Objetivo", y="Tempo de Processamento", data=df_final)
```

# Cálculo do algoritmo Busca em Profundidade com m = 10 #

```python
from main import profundidade
data_profundidade = profundidade()

lista_tempos = []
lista_achou= []
for tempos in data_profundidade["Tempo de Processamento"][:50]:
    
    if tempos != 0:
        lista_tempos.append(list(tempos)[0].microseconds)
        lista_achou.append(1)
    else:
        lista_tempos.append(0)
        lista_achou.append(0)

df_final_m10 = {
    "Algoritmo": data_profundidade["Algoritmo"][:50],
    "Objetivo": data_profundidade["Objetivo"][:50],
    "Tempo de Processamento": lista_tempos,
    "Achou": lista_achou
}

df_final_profundidade_m10 = pd.DataFrame(df_final_m10)
df_final_profundidade_m10.to_csv("profundidade_m10.csv")
df_final_profundidade_m10.head()
```

# Gráfico de tempo de processamento em função do número objetivo do Busca em Profundidade com m = 10 #

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.lineplot(x="Objetivo", y="Tempo de Processamento", data=df_final_profundidade_m10)
```

# Cálculo do algoritmo Busca em Profundidade com m = 100 #

```python
from main import profundidade
data_profundidade = profundidade()

lista_tempos = []
for tempos in data_profundidade["Tempo de Processamento"][50:]:
    if tempos != "...":
        lista_tempos.append(tempos.microseconds)
    else:
        lista_tempos.append(0)


df_final_m100 = {
    "Algoritmo": data_profundidade["Algoritmo"][50:],
    "Objetivo": data_profundidade["Objetivo"][50:],
    "Tempo de Processamento": lista_tempos
}


df_final_profundidade_m100 = pd.DataFrame(df_final_m100)
df_final_profundidade_m100.to_csv("profundidade_m100.csv")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
sns.lineplot(x="Objetivo", y="Tempo de Processamento", data=df_final_profundidade_m100)
```

# Cálculo do algoritmo Busca Profundidade Iterativa #

```python
from main import profundidade_iterativa
data_profundidade_iterativa = profundidade_iterativa()

```

```python
import pandas as pd
lista_tempos = []
for tempo in data_profundidade_iterativa["Tempo de Processamento"]:
    if tempo == "...":
        lista_tempos.append(0)
    else:
        lista_tempos.append(list(tempo)[0].microseconds)


df_final_profundidade_iterativa = {
    "Algoritmo": data_profundidade_iterativa["Algoritmo"],
    "Objetivo": data_profundidade_iterativa["Objetivo"],
    "Tempo de Processamento": lista_tempos
}

df_profundidade_iterativa = pd.DataFrame(df_final_profundidade_iterativa)
df_profundidade_iterativa.to_csv("profundidade_iterativa.csv")
```
